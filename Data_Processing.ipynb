```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Name": ["John", "Mary", "Peter", "Linda", None],
    "Age": [25, 30, np.nan, 28, 22],
    "Department": ["IT", "HR", "Finance", None, "IT"],
    "Salary": [50000, 60000, 55000, 52000, None]
}

df = pd.DataFrame(data)
df
```

```python
# Drop rows with null values
df_dropped = df.dropna()

# Drop column
df_no_salary = df.drop(columns=["Salary"])

df_dropped, df_no_salary
```

```python
df_filled = df.fillna({
    "Name": "Unknown",
    "Age": df["Age"].mean(),
    "Department": "Not Assigned",
    "Salary": df["Salary"].mean()
})

df_filled
```

```python
grouped = df_filled.groupby("Department")

grouped.describe()
```

```python
df2 = pd.DataFrame({
    "Name": ["Alice", "Bob"],
    "Age": [26, 32],
    "Department": ["IT", "HR"],
    "Salary": [48000, 62000]
})

concat_df = pd.concat([df_filled, df2])
concat_df
```

```python
dept_info = pd.DataFrame({
    "Department": ["IT", "HR", "Finance"],
    "Location": ["New York", "London", "Toronto"]
})

merged_df = pd.merge(df_filled, dept_info, on="Department", how="left")
merged_df
```
